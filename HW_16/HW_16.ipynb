## 1. Проверяем подключение к сайту гидрометцентра

```python
import requests

html = requests.get("https://meteoinfo.ru/forecasts/russia/republic-tatarstan")

if html.status_code == 200:
    print("Connection established successfully!")
else:
    print(f"Connection failed with status code {html.status_code}")
```

## 2. Путем хитрых манипуляций получаем список ссылок на текущую погоду во всех столицах регионов России

```python
from bs4 import BeautifulSoup
import re

soup = BeautifulSoup(html.content, 'html.parser')
location = soup.find("div", class_="jm-module-content clearfix")
region = location.find("div", id="div_2_2").find_all('option') # выбираем все сожержимое где есть полуссылки на регионы россии

regions = [] # список в строковом формате с названием регионов и полуссылкой
for n in region:
  regions.append(n.prettify())
regions
```

```python
# Полученный список преобразуем в одну строку
s =''
for n in regions:
  s = s + n
```

```python
# Регулярное выражение для поиска подстроки между 'value="russia/' и '">'
pattern = r"(?<=value=\"russia/)(.*?)(?=\">)"

reg = re.findall(pattern, s)
reg # список который хранит часть названия ссылки
```

```python
begin_http = "https://meteoinfo.ru/forecasts/russia/"

for (index, elem) in enumerate(reg):
  reg[index] = begin_http + reg[index]
reg # теперь список хранит список всех ссылок на столицы регионов россии.
```

## 3. Получаем список регионов в том же порядке как и список ссылко на регионы

```python
cities = []

for n in region:
  cities.append(n.text)
cities
```

## 4. Получаем по каждому региону данные в виде четырех списоков: температура, осадки, скорость ветра и давление на сегодняшний день

```python


temperature = []
precipitation = []
wind = []
pressure = []

for n in reg:
  html_city = requests.get(n)
  soup_city = BeautifulSoup(html_city.content, 'html.parser')
  temp = soup_city.find("div", class_="hidden-desktop")

  # извлекаем темепратуру
  try:
    data = temp.find('table').find('span', class_="fc_temp_short")
    temperature.append(data.text)
  except:
    temperature.append('--')

  # извлекаем осадки, скорость ветра и давление
  try:
    data_1 = temp.find('table').find_all('tr')
    data_2 = data_1[2].find('table').find('tr').find_all('td')

    precipitation.append(data_2[0].text)
    wind.append(data_2[1].text)
    pressure.append(data_2[2].text)
  except:
    precipitation.append('--')
    wind.append('--')
    pressure.append('--')

pressure
```

## 5. Конвертируем списки в dataframe

```python
import pandas as pd

intermediat_dictionary = {'Регион': cities,
                          'Температура': temperature,
                          'Осадки_мм': precipitation,
                          'СкоростьВетра_м/с': wind,
                          'Давление': pressure}

df = pd.DataFrame(intermediat_dictionary)
df
```
