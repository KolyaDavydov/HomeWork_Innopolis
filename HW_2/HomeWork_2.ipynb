```python
# Гипотеза:
# Введение санкций и уход более 40 инностранных автопроизводителей в апреле 2022 г.
# увеличит производство и продажу новых автомобилей "ЛАДА" на 10%

# Берем статистику продаж новых автомобилей "ЛАДА" с сайта https://abreview.ru/stat/aeb/

# событие: уход иностранных производителей авто
# Данные продаж по месяцам до события (апрель 21г - март 22 г)

before_21 = {'апрель': 37931,
             'май': 38284,
             'июнь': 40096,
             'июль': 32013,
             'август': 19428,
             'сентябрь': 17189,
             'октябрь': 25573,
             'ноябрь': 26373,
             'декабрь': 29919,
             'январь': 18045,
             'февраль': 22378,
             'март': 12289}
# Данные продаж по месяцам после события (апрель 22г - март 23 г)
after_21 = {'апрель': 8506,
             'май': 6012,
             'июнь': 7484,
             'июль': 10323,
             'август': 18087,
             'сентябрь': 20641,
             'октябрь': 19145,
             'ноябрь': 21726,
             'декабрь': 24009,
             'январь': 17468,
             'февраль': 23437,
             'март': 23335}

# продано за год до события:
avaregeCar_21 = 0

for value in before_21.values():
  avaregeCar_21 += value
print("До события проданно: " + str(avaregeCar_21))

# продано за год после события:
avaregeCar_22 = 0

for value in after_21.values():
  avaregeCar_22 += value
print("После события проданно: " + str(avaregeCar_22))

# ПРОВЕРКА ГИПОТЕЗЫ

#доверительный интервал %
p = 0.05
print("Доверительный интервал: " + str(p * 100) + "% от 10%")

gip = avaregeCar_21 * 1.1

if avaregeCar_22 > (gip * (1 + p)) or avaregeCar_22 < (gip * (1 - p)):
  print("Гипотеза отклонена")
else:
  print("Гипотеза не отклонена")
```
